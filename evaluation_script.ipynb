```python
import numpy,utils
import numpy as np
import matplotlib
from matplotlib import pyplot as plt
%matplotlib inline
from IPython.display import set_matplotlib_formats
set_matplotlib_formats('pdf','png')
plt.rcParams['savefig.dpi'] = 90

dlist= numpy.arange(0,100)  # Number of Epochs

#This function extrat the data from the log
#it return: speedlist, accuracylist of the training.
# and 
def extract_data(filename):
    inFile = open(filename)
    #training Data
    accuracylist = np.array([]) # Accuracy pro Epoch 
    speedlist = np.array([])    # Speed of one Epoch (image/sec)
    
    #To-Do: extract number of Epochs
    #dlist= numpy.arange(0,552)      # Number of Epochs

    #Validation Data
    val_acc_list = np.array([])      # Accuracy  
    val_speed_list = np.array([])    # Speed (image/sec)
   
    buffer = []
    for line in inFile:
        buffer.append(line)
        if line.startswith("Epoch["):
            word = line.split("\t")
            #print(word)
            speedString = word[1].split(" ")[1]
            speedlist = np.append(speedlist , [speedString])
            accuracyString = word[2].split("=")[1]
            accuracylist = np.append(accuracylist , [accuracyString])
        #For extracting validation Data
        #elif line.startswith(""):
            #print("no epoch" , line )
    accuracylist = accuracylist.astype(np.float)
    accuracylist = np.round(accuracylist, 2)
    
    speedlist = speedlist.astype(np.float)
    #accuracylist = np.round(accuracylist, 2)
    #print(accuracylist[0:100])
    inFile.close()
    #outFile.close()
    return speedlist, accuracylist

lists = extract_data("mxnet.log")
print(lists)


'''
graph = plt.figure(figsize=(30, 30))
plt.plot(dlist, lists[0][0:100],'g^', label='Mxnet', linewidth=3.0, markersize=20)
axis_font = {'fontname':'Arial', 'size':'14'}
plt.title('Speed'); plt.xlabel('Epochs', fontsize=30); plt.ylabel('Speed', fontsize=30)
graph.savefig("Speed.jpg", bbox_inches='tight')

graph = plt.figure(figsize=(30, 30))
plt.plot(dlist, lists[1][0:100],'bo', label='Mxnet', linewidth=2.0, markersize=20)
plt.title('Accuracy'); plt.xlabel('Epochs' , fontsize=30); plt.ylabel('Acc', fontsize=30)
graph.savefig("Accuracy.jpg", bbox_inches='tight')
'''
```
