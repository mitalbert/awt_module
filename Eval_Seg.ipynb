```python
import numpy as np
import re


# This function extract the data from the log
# it return: accuracy list of the testing,loss list of the testing, loss list of the training
def extract_data(filename):

    # training Data
    loss_list = np.array([])  # loss list of the training

    # testing Data
    accuracy_list = np.array([])  # accuracy list of the testing

    for line in open(filename):
        match_loss_test = re.search(r'.* Loss: (\d*\.*\d*)\n', line)
        match_acc_test = re.search(r'.* Mean IoU : 	: (\d*\.*\d*)\n', line)

        if match_acc_test:
            accuracy_list = np.append(accuracy_list, float(match_acc_test.group(1)))
            #print ("Accu",match_acc_test.group(1))

        elif match_loss_test:
            loss_list = np.append(loss_list, float(match_loss_test.group(1)))
            #print ("Loss",match_loss_test.group(1))


    return accuracy_list, loss_list


accuracy, loss = extract_data("segmentation_from_scratch_run.log")

print("accuracy list\n", accuracy)
print("loss list\n", loss)

```

```python
import numpy
import matplotlib
from matplotlib import pyplot as plt
%matplotlib inline
from IPython.display import set_matplotlib_formats
```

```python
dlist = numpy.arange(0,accuracy.size)

set_matplotlib_formats('pdf','png')
plt.rcParams['savefig.dpi'] = 90

graph = plt.figure()
graph.set_figheight(30)
graph.set_figwidth(30)
plt.plot(dlist,accuracy ,'--', markersize=20, color = "blue" )


plt.xlabel('Epochs' , fontsize=50); 
plt.ylabel('Accuracy', fontsize=50)
#plt.legend(prop={'size': 60}, loc = 'lower right')
plt.xticks(fontsize=30)
plt.yticks(fontsize=30)
plt.autoscale(enable=True, axis='x')
#plt.title('Accuracy' , fontsize=50)
plt.show()
graph.savefig("Accuracy.jpg")
```
